```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

salary_data = pd.read_csv('salaryData.csv')
```

```python
salary_data.columns
```

```python
salary_data.tail()
```

```python
sns.scatterplot(x='yearsofexperience', y='totalyearlycompensation', data=salary_data, hue='gender')
```

```python
keywords = [ 'software', 'principle', 'manager', 'engineer']
salary_data_subset = salary_data[salary_data['title']=='Software Engineer']
```

```python
sns.scatterplot(x='yearsofexperience', y='totalyearlycompensation', data=salary_data_subset, hue='gender')
```

```python
salary_data_subset = salary_data[salary_data['title']=='Software Engineer']
sns.displot(x='yearsofexperience', y='totalyearlycompensation', kind='kde', data=salary_data_subset)
```

```python
df = salary_data_subset.where(salary_data_subset['yearsofexperience'] <= 20).groupby('yearsofexperience', as_index=False).totalyearlycompensation.median()

df = df.rename(columns={'totalyearlycompensation': 'compensation'})

f,ax = plt.subplots(figsize=(10,10))

sns.lineplot(x='yearsofexperience', y='compensation', data=df)
```

```python
f, ax = plt.subplots(figsize=(15,10))
ax.set_xscale('log')
sns.boxplot(x='totalyearlycompensation', y='title', data=salary_data)
sns.stripplot(x="totalyearlycompensation", y="title", data=salary_data,
              size=1, color=".3", linewidth=0)
ax.set(ylabel="")
ax.xaxis.grid(True)
```

```python
cisco_data = salary_data[salary_data['company']=='Cisco']
cisco_data.tail()
```

```python
f, ax = plt.subplots(figsize=(15,10))
ax.set_xscale('log')
sns.boxplot(x='totalyearlycompensation', y='title', data=cisco_data)
sns.stripplot(x="totalyearlycompensation", y="title", data=cisco_data,
              size=1, color=".3", linewidth=0)
ax.set(ylabel="")
ax.xaxis.grid(True)
```

```python
f, ax = plt.subplots(figsize=(10,10))
sns.scatterplot(x='yearsofexperience', y='totalyearlycompensation', hue='yearsatcompany', data=cisco_data)
```

```python
f, ax = plt.subplots(ncols=2, figsize=(20,10))

ax[0].set_title('Global')
ax[1].set_title('Cisco')

sns.scatterplot(ax=ax[0], x='yearsofexperience', y='yearsatcompany', data=salary_data)
sns.scatterplot(ax=ax[1], x='yearsofexperience', y='yearsatcompany', data=cisco_data)
```

```python
f, ax = plt.subplots(ncols=2, figsize=(20,10))

ax[0].set_title('Global')
ax[1].set_title('Cisco')

sns.heatmap(salary_data.corr(), ax=ax[0])
sns.heatmap(cisco_data.corr(), ax=ax[1])
```

```python
f, ax = plt.subplots(ncols=2, figsize=(20,10))

ax[0].set_title('Global')
ax[1].set_title('Cisco')

d1 = salary_data[salary_data['title'] == 'Software Engineer']
d2 = cisco_data[cisco_data['title'] == 'Software Engineer']

sns.heatmap(d1.corr(), ax=ax[0])
sns.heatmap(d2.corr(), ax=ax[1])
```

```python
f, ax = plt.subplots(figsize=(12,8))

#sns.displot(x='totalyearlycompensation', col='title', row='location', data=cisco_data)
sns.histplot(x='totalyearlycompensation', kde=True, data=cisco_data[(cisco_data.title == 'Software Engineer') 
                                                          & (cisco_data.location == 'San Jose, CA')
                                                          & (cisco_data.yearsofexperience < 5)])
```
